```
import glob

def create_test_model():
  #set parameters
  set_hyperparameters(baseline="PPO")
  h.value_clipping = True
  h.version = "Load_and_save_test"
  h.total_steps = 2e5

  #create model and train for a bit
  model = PPO(print_output=True, eval=True, save_interval = 5e4)
  file_name = model.file_name
  model.train()

def load_test_model(file_name):
  #set parameters
  set_hyperparameters(baseline="PPO")
  h.value_clipping = True
  h.version = "Load_and_save_test"
  h.total_steps = 2e5

  #create model and train for a bit
  model = PPO(print_output=True, eval=True, save_interval = 5e4)
  model.load_policy(file_name)
  return model

def print_all_related_models(file_name):
  all_models = glob.glob(MODEL_PATH +'*.pt')
  related_models = []
  for model in all_models:
    if file_name in model:
      related_models.append(model)
  print("Related models:")
  for model in related_models:
    print(model)
```
